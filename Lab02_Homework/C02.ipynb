```python
def fibonacci(upper_threshold: int) -> list:
  fibb = [0 , 1]
  i = 2
  while max(fibb) < upper_threshold:
    if (fibb[i-2] + fibb[i-1]) > upper_threshold:
      break
    else:
      fibb.append(fibb[i-2] + fibb[i-1])
    i += 1
  return fibb

print(fibonacci(34))
```

```python
import numpy as np
import time

def create_array_nxn(n: int) -> np.ndarray:
  arr = np.arange(n*n)[::-1]
  arr = arr.reshape((n,n))
  return arr

def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    arr_shape = arr.shape
    arr = arr.reshape(arr.size,1)
    for i in range(arr.size):
      if arr[i] > threshold:
        arr[i] = 0
    arr = arr.reshape(arr_shape)
    return arr

def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
  arr = np.where(arr < threshold, arr, 0)
  return arr

def compare_performance(n: int, threshold: int) -> None:
  arr = np.arange(n*n)[::-1]
  arr = arr.reshape((n,n))

  start_time_loop = time.time()
  apply_threshold_loop(arr,threshold)
  stop_time_loop = time.time()

  start_time_vector = time.time()
  apply_threshold_vectorized(arr,threshold)
  stop_time_vector = time.time()

  time_difference_loop = stop_time_loop - start_time_loop
  time_difference_vector = stop_time_vector - start_time_vector

  print("Execution time of loop based thresholding is:", time_difference_loop)
  print("Execution time of vectorised based thresholding is:", time_difference_vector)

  return
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load
dataset = pd.read_csv("sample_data/california_housing_train.csv")

# 1) describe
print(dataset.describe())

# 2) rows where total_bedrooms > 310
print(dataset[dataset["total_bedrooms"] > 310])

# 3) drop first + last row
dataset = dataset.iloc[1:-1]

# 4) households mean + dot plot
print("Mean households:", dataset["households"].mean())
plt.scatter(dataset.index, dataset["households"], s=10)
plt.show()

# 5) fill NaNs with column mean
dataset = dataset.fillna(dataset.mean())

# 6) lat vs long plot
plt.scatter(dataset["latitude"], dataset["longitude"], s=10)
plt.show()

# 7) min-max normalization (2 columns)
for col in ["total_rooms", "median_income"]:
    dataset[col+"_norm"] = (dataset[col] - dataset[col].min()) / (dataset[col].max() - dataset[col].min())

# 8) correlation matrix
corr = dataset.corr()
print(corr)
plt.imshow(corr)
plt.colorbar()
plt.show()
```
